```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('dataset/Final_data.csv')
df.info()
df
```

```python
df.isna().sum()
```

```python
cols_to_drop = ['Benefit', 'Workout', 'BMI', 'BMI_calc', 'cal_from_macros',
                'pct_carbs', 'protein_per_kg', 'pct_maxHR', 'cal_balance',
                'lean_mass_kg', 'expected_burn', 'Burns Calories (per 30 min)_bc',
                'Burns_Calories_Bin', 'meal_name']
df.drop(columns=cols_to_drop, inplace=True)
print(f"columns dropped. new shape: {df.shape}")
```

```python
mask = (df['meal_type'] != 'Other') & \
           (df['diet_type'] != 'Other')
    
df = df[mask]
print(f"Shape after filtering 'Other': {df.shape}")
```

```python
float_cols = [
        'Weight (kg)', 'Height (m)', 'Max_BPM', 'Avg_BPM', 'Resting_BPM', 
        'Session_Duration (hours)', 'Calories_Burned', 'Fat_Percentage', 
        'Water_Intake (liters)', 'Carbs', 'Proteins', 'Fats', 'Calories', 
        'sugar_g', 'sodium_mg', 'cholesterol_mg', 'serving_size_g', 
        'prep_time_min', 'cook_time_min', 'rating', 'Burns Calories (per 30 min)'
    ]
    
int_cols = [
    'Age', 'Workout_Frequency (days/week)', 'Experience_Level', 
    'Daily meals frequency', 'Sets', 'Reps'
]

existing_float_cols = [col for col in float_cols if col in df.columns]
existing_int_cols = [col for col in int_cols if col in df.columns]

df[existing_float_cols] = df[existing_float_cols].round(2)
df[existing_int_cols] = df[existing_int_cols].round(0).astype(int)
print("Applied data type corrections.")
```

```python
output_dir = 'dataset/'
os.makedirs(output_dir, exist_ok=True)

output_path = os.path.join(output_dir, 'cleaned_data.csv')
df.to_csv(output_path, index=False)
```
